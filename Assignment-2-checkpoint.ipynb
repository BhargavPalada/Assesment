```python
import pandas as pd

# Load CSV
df_csv = pd.read_csv("etrm_trades/etrm_trades.csv")

# Load JSON
df_json = pd.read_json("etrm_trades/etrm_trades.json")

# Load Excel
df_xlsx = pd.read_excel("etrm_trades/etrm_trades.xlsx")

# Load TXT (assuming comma-separated, change sep if needed)
df_txt = pd.read_csv("etrm_trades/etrm_trades.txt", sep=",")

# Load HTML (returns a list of DataFrames)
df_html_list = pd.read_html("etrm_trades/etrm_trades.html")

# Load XML
df_xml = pd.read_xml("etrm_trades/etrm_trades.xml")
```

```python
print(df_csv.dtypes)
```

```python
#Converting object datatype into datetime
df_csv['DeliveryStart']=pd.to_datetime(df_csv['DeliveryStart'])
df_json['DeliveryStart']=pd.to_datetime(df_json['DeliveryStart'])
df_xlsx['DeliveryStart']=pd.to_datetime(df_xlsx['DeliveryStart'])
df_txt['DeliveryStart']=pd.to_datetime(df_txt['DeliveryStart'])
df_html_list[0]['DeliveryStart']=pd.to_datetime(df_html_list[0]['DeliveryStart'])
df_xml['DeliveryStart']=pd.to_datetime(df_xml['DeliveryStart'])

df_csv['DeliveryEnd']=pd.to_datetime(df_csv['DeliveryEnd'])
df_json['DeliveryEnd']=pd.to_datetime(df_json['DeliveryEnd'])
df_xlsx['DeliveryEnd']=pd.to_datetime(df_xlsx['DeliveryEnd'])
#df_txt['DeliveryEnd']=pd.to_datetime(df_txt['DeliveryEnd'])
df_html_list[0]['DeliveryEnd']=pd.to_datetime(df_html_list[0]['DeliveryEnd'])
df_xml['DeliveryEnd']=pd.to_datetime(df_xml['DeliveryEnd']) 
```

```python
#Checking for datatype
print(df_csv.dtypes)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


# Distribution of 'Price' column
plt.figure(figsize=(10, 6))
sns.histplot(df_csv['Price'], bins=30, kde=True)
plt.title('Distribution of Price')
```

```python
#Analysis on Volume done by buying and selling on various commodity
grouped = df_csv.groupby(["Commodity", "TradeType"])["Volume"].sum().unstack(fill_value=0)

print("Trade Volume by Commodity and Trade Type:\n")
print(grouped)

grouped.plot(kind="bar", stacked=True, figsize=(10,6), colormap="tab20")

plt.title("Trade Volume by Commodity and Trade Type")
plt.xlabel("Commodity")
plt.ylabel("Total Volume")
plt.legend(title="Trade Type")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_traders = df_csv.groupby("TradeID")["Volume"].sum().sort_values(ascending=False).head(5)
print("Top 5 Traders by Trade Volume:\n", top_traders, "\n")

# Plot on top 5 Trader by trade volume
top_traders.plot(kind="bar", color="skyblue", edgecolor="black", figsize=(7,5))
plt.title("Top 5 Traders by Trade Volume")
plt.ylabel("Total Volume")
plt.xlabel("TradeID")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
avg_price = df_csv.groupby("Commodity")["Price"].mean()
print("Average Price per Commodity:\n", avg_price, "\n")

# Plot on average price per commodity
avg_price.plot(kind="bar", color="lightgreen", edgecolor="black", figsize=(7,5))
plt.title("Average Price per Commodity")
plt.ylabel("Average Price")
plt.xlabel("Commodity")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
if "Currency" in df_html_list[0].columns:
    currency_dist = df_html_list[0]["Currency"].value_counts()
    print("Distribution of Trades by Currency:\n", currency_dist, "\n")

    # Plot on distribution of trades by Currency
    currency_dist.plot(kind="pie", autopct="%1.1f%%", figsize=(6,6), startangle=90, colormap="Set3")
    plt.title("Distribution of Trades by Currency")
    plt.ylabel("")  # hide ylabel
    plt.tight_layout()
    plt.show()
```

```python
if "Periodicity" in df_csv.columns:
    periodicity_breakdown = df_csv["Periodicity"].value_counts()
    print("Trade Periodicity Breakdown:\n", periodicity_breakdown, "\n")

    # Plot on trade periodicity breakdown
    periodicity_breakdown.plot(kind="bar", color="salmon", edgecolor="black", figsize=(7,5))
    plt.title("Trade Periodicity Breakdown")
    plt.ylabel("Number of Trades")
    plt.xlabel("Periodicity")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
#Plot on Trade Volume per commodity
commodity_vol = df_csv.groupby("Commodity")["Volume"].sum()
commodity_vol.plot(kind="bar", color="skyblue", edgecolor="black", figsize=(6,4))
plt.title("Total Volume per Commodity")
plt.ylabel("Total Volume")
plt.xlabel("Commodity")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Plot on price distribution
df_csv["Price"].plot(kind="hist", bins=10, color="salmon", edgecolor="black", figsize=(6,4))
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
#Plot on Average Price Trend by Delivery Start Date
df_xml['DeliveryStart'] = pd.to_datetime(df_xml['DeliveryStart'])

# Group by delivery date and calculate average price
avg_price_trend = df_xml.groupby('DeliveryStart')['Price'].mean()

# Plot line chart
plt.figure(figsize=(10, 6))
plt.plot(avg_price_trend.index, avg_price_trend.values, marker='o', linestyle='-')
plt.title("Average Price Trend by Delivery Start Date", fontsize=14)
plt.xlabel("Delivery Start Date", fontsize=12)
plt.ylabel("Average Price", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.xticks(rotation=45)

plt.show()
```

```python
#Plot on Distribution of Notional Values
plt.figure(figsize=(10, 6))
df_csv['Notional']=df_csv['Volume']*df_csv['Price']
plt.hist(df_csv['Notional'], bins=30, edgecolor='black', alpha=0.7)

plt.title("Distribution of Notional Values", fontsize=14)
plt.xlabel("Notional Value", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle="--", alpha=0.6)

plt.show()
```

```python
pivot = pd.pivot_table(df_csv, values="TradeID", index="Commodity", columns="Currency", aggfunc="count", fill_value=0)
 
# Plot heatmap on Trade by Commodity and Currency
plt.figure(figsize=(8,6))
sns.heatmap(pivot, annot=True, fmt="d", cmap="Blues", cbar=True)

plt.title("Heatmap of Trades by Commodity and Currency", fontsize=14)
plt.xlabel("Currency")
plt.ylabel("Commodity")
plt.show()
```
